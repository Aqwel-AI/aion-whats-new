## 2. New Package: `aion.algorithms`

The **algorithms** package provides classical search, array, and graph utilities for research and education. Implementations prioritize correctness, clarity, and predictable behavior.

### 2.1 Search algorithms (`aion.algorithms.search`)

| Function | Description | Complexity |
|----------|-------------|------------|
| `binary_search(arr, target)` | Find target in sorted list; returns index or `None` | O(log n) |
| `lower_bound(arr, target)` | First index where element ≥ target (insertion point) | O(log n) |
| `upper_bound(arr, target)` | First index where element > target | O(log n) |
| `is_sorted(ulist)` | Check if list is sorted ascending or descending | O(n) |
| `jump_search(slist, step, target)` | Jump by `step` then linear search in block | O(√n) |
| `find_peak_element(ulist)` | Return all elements strictly greater than neighbors | O(n) |
| `exponential_search(slist, target)` | Expand range exponentially then binary search | O(log n) |
| `linear_search(ulist, target)` | Sequential search; returns message string | O(n) |
| `First_Occurrence(ulist, target)` | Index of first occurrence of target | O(n) |
| `Last_Occurrence(ulist, target)` | Index of last occurrence of target | O(n) |
| `First_Last_Occurrence(ulist, target)` | Tuple (first, last) occurrence | O(n) |
| `roatated_search(ulist, target)` | Search in rotated sorted array | O(log n) |
| `ternary_search(slist, target)` | Divide into three segments | O(log₃ n) |
| `interpolation_search(slist, target)` | Estimate position for uniform data | O(log log n) avg |

**Parameters & returns (summary):**
- **`binary_search(arr, target)`** — `arr`: sorted list; returns `int | None`. No exceptions.
- **`lower_bound(arr, target)`** — First index where `arr[i] >= target`; else `len(arr)`.
- **`upper_bound(arr, target)`** — First index where `arr[i] > target`; else `len(arr)`.
- **`is_sorted(ulist)`** — True if non-decreasing or non-increasing. **Raises:** `TypeError` if not a list.
- **`jump_search(slist, step, target)`** — `step` must be positive. **Raises:** `TypeError`, `ValueError`.
- **`find_peak_element(ulist)`** — List of peak values. **Raises:** `TypeError` if not list, `ValueError` if `len < 2`.
- **`linear_search`, `First_Occurrence`, `Last_Occurrence`** — Return string (index or "not found"). **Raises:** `TypeError` if invalid types.
- **`roatated_search`, `ternary_search`, `interpolation_search`** — Return string. **Raises:** `TypeError`; interpolation expects numeric sorted data.

**When to use which:** Use **binary_search** for exact match on sorted data; **lower_bound** / **upper_bound** for range queries or insertion point; **exponential_search** when the list is huge or unbounded; **interpolation_search** for uniformly distributed sorted numbers; **jump_search** when you want fixed-size blocks; **linear_search** / **First_Occurrence** / **Last_Occurrence** for unsorted or when you need first/last index; **roatated_search** for rotated sorted arrays; **find_peak_element** for 1D peak finding.

````python
from aion.algorithms import (
    binary_search,
    lower_bound,
    upper_bound,
    is_sorted,
    jump_search,
    find_peak_element,
    exponential_search,
    linear_search,
    First_Occurrence,
    Last_Occurrence,
    First_Last_Occurrence,
    roatated_search,
    ternary_search,
    interpolation_search,
)

# Sorted list for search examples
arr = [1, 2, 4, 4, 4, 6, 8, 10]

print("binary_search(arr, 4)     ->", binary_search(arr, 4))
print("binary_search(arr, 5)     ->", binary_search(arr, 5))
print("lower_bound(arr, 4)       ->", lower_bound(arr, 4), "(first index where val >= 4)")
print("upper_bound(arr, 4)       ->", upper_bound(arr, 4), "(first index where val > 4)")
print("is_sorted(arr)            ->", is_sorted(arr))
print("is_sorted([3,2,1])        ->", is_sorted([3, 2, 1]))
print("jump_search(arr, 3, 6)    ->", jump_search(arr, 3, 6))
print("find_peak_element([1,3,2,1]) ->", find_peak_element([1, 3, 2, 1]))
print("exponential_search(arr, 8)->", exponential_search(arr, 8))
print("linear_search(arr, 6)     ->", linear_search(arr, 6))
print("First_Occurrence(arr, 4)  ->", First_Occurrence(arr, 4))
print("Last_Occurrence(arr, 4)   ->", Last_Occurrence(arr, 4))
print("First_Last_Occurrence(arr, 4) ->", First_Last_Occurrence(arr, 4))
````

````python
# Example: count how many elements equal to target using lower_bound and upper_bound
arr = [1, 2, 4, 4, 4, 6, 8, 10]
target = 4
lo = lower_bound(arr, target)
hi = upper_bound(arr, target)
print(f"Count of {target}: {hi - lo}. Insert 5 at index {upper_bound(arr, 5)} to keep sorted.")
````

````python
# Rotated sorted array: [4,5,6,1,2,3]
rotated = [4, 5, 6, 1, 2, 3]
print("roatated_search(rotated, 2) ->", roatated_search(rotated, 2))
print("ternary_search(arr, 6)      ->", ternary_search(arr, 6))

# Interpolation search works best on uniformly distributed data
uniform = list(range(0, 100, 5))  # [0, 5, 10, ..., 95]
print("interpolation_search(uniform, 50) ->", interpolation_search(uniform, 50))
````

| Function | Description |
|----------|-------------|
| `flatten_array(arr)` | Flatten list of lists into one list (one level). |
| `flatten_deep(arr)` | Recursively flatten nested lists to any depth. |
| `chunk_array(arr, size)` | Split list into chunks of given `size`; last chunk may be smaller. |
| `remove_duplicates(arr)` | Remove duplicates while preserving order. |

| `pad_array(arr, min_len, item)` | Extend list to `min_len` by appending `item`. |
| `rolling_sum(arr, size)` | List of sums of each consecutive window of length `size`. |
| `pairwise(arr)` | List of consecutive pairs `(arr[i], arr[i+1])`. |

````python
from aion.algorithms import (
    flatten_array,
    chunk_array,
    remove_duplicates,
    moving_avarage,
    flatten_deep,
    sliding_window,
    pad_array,
    rolling_sum,
    pairwise,
)

# Flatten
print("flatten_array([[1,2],[3,4]])     ->", flatten_array([[1, 2], [3, 4]]))
print("flatten_deep([1,[2,[3,4]]])      ->", flatten_deep([1, [2, [3, 4]]]))

# Chunk and deduplicate
print("chunk_array([1,2,3,4,5], 2)      ->", chunk_array([1, 2, 3, 4, 5], 2))
print("remove_duplicates([1,2,1,3,2])   ->", remove_duplicates([1, 2, 1, 3, 2]))
print("moving_avarage([1.0, 2.0, 3.0]) ->", moving_avarage([1.0, 2.0, 3.0]))

# Sliding window (generator)
print("list(sliding_window([1,2,3,4], 2)) ->", list(sliding_window([1, 2, 3, 4], 2)))

# Padding (modifies list in place)
p = [1, 2]
print("pad_array([1,2], 5, 0)          ->", pad_array(p.copy(), 5, 0))

print("rolling_sum([1,2,3,4], 2)       ->", rolling_sum([1, 2, 3, 4], 2))
print("pairwise([1,2,3,4])             ->", pairwise([1, 2, 3, 4]))
````

### 2.3 Graph algorithms (`aion.algorithms.graphs`)

Graphs are represented as **adjacency lists**: `dict` mapping each node to a list of neighbors.

| Function | Description | Complexity |
|----------|-------------|------------|
| `bfs(graph, start)` | Breadth-first search from `start`; returns list of nodes in BFS order. | O(V+E) |
| `dfs(graph, start)` | Depth-first search (iterative, stack-based); returns nodes in pre-order. | O(V+E) |
| `toposort(graph)` | Topological sort of a DAG (Kahn's algorithm); raises if cycle detected. | O(V+E) |

````python
from aion.algorithms import bfs, dfs, toposort

# Undirected graph: 0 -- 1 -- 2
#                    \  /
#                     3
graph = {0: [1, 3], 1: [0, 2, 3], 2: [1], 3: [0, 1]}
print("BFS from 0:", bfs(graph, 0))
print("DFS from 0:", dfs(graph, 0))

# DAG for topological sort (e.g. task dependencies)
dag = {"A": ["B", "C"], "B": ["D"], "C": ["D"], "D": []}
print("Topological order:", toposort(dag))
````

**Graph representation & toposort**

- **Adjacency list:** `graph[node]` = list of neighbors (for directed: successors). Nodes must be **hashable**.
- **BFS / DFS:** **Raises** `KeyError` if `start` not in graph. Use `graph.setdefault(start, [])` if needed.
- **toposort(graph):** Expects a **DAG**. **Raises** `ValueError("Graph contains a cycle; topological sort is undefined")` if cycle.

````python
# toposort raises ValueError when graph has a cycle
from aion.algorithms import toposort
cyclic = {"A": ["B"], "B": ["C"], "C": ["A"]}
try:
    toposort(cyclic)
except ValueError as e:
    print("Cycle detected:", e)
````

## 5. Troubleshooting & FAQ

| Issue | Cause | Fix |
|-------|--------|-----|
| **`KeyError: 'start' not in graph`** | BFS/DFS called with `start` not in the graph. | Add the node: `graph.setdefault(start, [])` or ensure `start` is a key. |
| **`ValueError: Graph contains a cycle`** | `toposort()` was given a graph with a cycle. | Ensure the graph is a DAG, or catch the exception. |
| **`ValueError: List must not be empty`** | Empty list passed to flatten_array, chunk_array, etc. | Check `if not arr: ...` before calling. |
| **Wrong or missing search result** | List not sorted for binary_search, lower_bound, etc. | Use `is_sorted(arr)` to verify; for unsorted use linear_search or First/Last occurrence. |
| **Plot not showing** | Headless environment (CI, SSH). | Use `show=False` and `save_plot(fig, "out.png")` or `plt.close(fig)`. |
| **`pad_array` changed my list** | It modifies the list in place. | Use `pad_array(arr.copy(), min_len, item)`. |
| **`ModuleNotFoundError: aion`** | Package not installed or path not set. | `pip install -e "aion(v0.1.8)/"` or `sys.path.insert(0, "aion(v0.1.8)")`. |
| **Matplotlib missing** | Base install may not include plotting. | `pip install aqwel-aion[full]` or `pip install matplotlib numpy`. |

## 7. Aion vs standard library

| Need | Standard library | Aion (v0.1.8) |
|------|-------------------|----------------|
| Binary search / insertion point | `bisect.bisect_left`, `bisect_right` | `lower_bound`, `upper_bound` — same idea; use aion for one consistent API. |
| Exact match in sorted list | `bisect` + equality check | `binary_search(arr, target)` → index or None. |
| Consecutive pairs | `itertools.pairwise` (3.10+) | `pairwise(arr)` → list of tuples. |
| Chunking | Manual loop or more_itertools | `chunk_array(arr, size)`. |
| BFS / DFS / toposort | Implement or networkx | `bfs`, `dfs`, `toposort` — no extra deps. |

## 8. End-to-end workflow example

Combine **algorithms** and **visualization**: synthetic data → rolling sum → range query (lower/upper_bound) → plot.

````python
# Workflow: signal → rolling_sum → lower_bound/upper_bound → plot
import numpy as np
from aion.algorithms import rolling_sum, lower_bound, upper_bound
from aion.visualization import plot_multiple_arrays
import matplotlib.pyplot as plt

np.random.seed(42)
signal = np.cumsum(np.random.randn(80)) + 20
rolling = rolling_sum(signal.astype(int).tolist(), size=5)
sorted_rolling = sorted(rolling)
lo, hi = lower_bound(sorted_rolling, 100), upper_bound(sorted_rolling, 200)
fig = plot_multiple_arrays(
    [signal.tolist(), rolling + [0] * (len(signal) - len(rolling))],
    labels=["Signal", "Rolling sum"],
    title="Signal + rolling sum (aion workflow)",
    show=False,
)
plt.close(fig)
print("Rolling sum count:", len(rolling), "| Values in [100,200]:", hi - lo)
````
